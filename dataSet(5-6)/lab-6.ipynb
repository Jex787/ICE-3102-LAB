```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/car-driving-risk-analysis/car driving risk analysis.csv')
df.head() #display first 5 row
```

```python
x = df[['speed']]
y = df[['risk']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.4, random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() #create object
```

```python
reg.fit(xtrain, ytrain)
```

```python
reg.predict(xtest)
```

```python
ytest
```

```python
plt.scatter(df['speed'], df['risk'], marker='*', color='red')
plt.xlabel('Speed of Car')
plt.ylabel('Car Driving Speed risk')
plt.plot(df.speed, reg.predict(df[['speed']]))
# plt.show()
```
